```python
#using the eblow method
wss=[]
for i in range(1,11):
    k=KMeans(n_clusters=i,init='k-means++',random_state=123)
    k.fit(df)
    wss.append(k.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wss,marker='8',color='red',markerfacecolor='green')
plt.title("Elbow Method")
plt.xlabel("Clusters")
plt.ylabel("WSS")
plt.show()
```
